```python
# -------------------------------
# 1. Import Libraries
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn theme
sns.set_theme(style="whitegrid")  


# -------------------------------
# 2. Load Datasets
# -------------------------------
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
gender = pd.read_csv("gender_submission.csv")

print("Train Shape:", train.shape)
print("Test Shape:", test.shape)
print("Gender Submission Shape:", gender.shape)

# Merge gender_submission with test to get Survived column (optional)
test_full = test.merge(gender, on="PassengerId", how="left")

# -------------------------------
# 3. Basic Exploration
# -------------------------------
print(train.head())
print(train.info())
print(train.describe())

# Missing values
print("Missing values in Train:\n", train.isnull().sum())

# -------------------------------
# 4. Univariate Analysis
# -------------------------------

# Survival counts
sns.countplot(x="Survived", data=train, palette="Set2")
plt.title("Survival Counts")
plt.show()

# Passenger Class
sns.countplot(x="Pclass", data=train, palette="Set3")
plt.title("Passenger Class Distribution")
plt.show()

# Gender distribution
sns.countplot(x="Sex", data=train, palette="Set1")
plt.title("Gender Distribution")
plt.show()

# Age Distribution
sns.histplot(train["Age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

# -------------------------------
# 5. Bivariate Analysis
# -------------------------------

# Survival by Gender
sns.countplot(x="Survived", hue="Sex", data=train, palette="coolwarm")
plt.title("Survival by Gender")
plt.show()

# Survival by Class
sns.countplot(x="Survived", hue="Pclass", data=train, palette="viridis")
plt.title("Survival by Class")
plt.show()

# Survival by Embarked Port
sns.countplot(x="Survived", hue="Embarked", data=train, palette="Set2")
plt.title("Survival by Embarkation Port")
plt.show()

# -------------------------------
# 6. Correlation Analysis
# -------------------------------
corr = train.corr(numeric_only=True)
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# -------------------------------
# 7. Advanced Plots
# -------------------------------

# Boxplot: Age vs Survival
sns.boxplot(x="Survived", y="Age", data=train, palette="Set2")
plt.title("Age vs Survival")
plt.show()

# Scatterplot: Fare vs Age
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=train, palette="deep")
plt.title("Fare vs Age (Survival Highlighted)")
plt.show()

# -------------------------------
# 8. Key Observations
# -------------------------------
print("Observations:")
print("- More passengers did not survive (0) than survived (1).")
print("- Higher survival rates for females compared to males.")
print("- 1st Class passengers had higher survival rates than 2nd/3rd.")
print("- Most passengers were in 3rd class.")
print("- Younger passengers and higher fare-paying passengers had better survival chances.")
print("- Embarked port 'C' had higher survival compared to 'Q' and 'S'.")
```
